```python
import json
import traceback

from fabrictestbed_extensions.fablib.fablib import fablib
```

```python
SLICE_NAME = 'p4-simple-router'

# We will use Ubuntu 20.04 for both nodes
IMAGE = 'default_ubuntu_20'
CORES = 1
RAM = 2

site1 = 'UTAH'
site2 = 'TACC'
site3 = 'SALT'
```

```python
import datetime
from datetime import timedelta

try:
    #Create Slice
    slice = fablib.new_slice(name=SLICE_NAME)
    #Create P4 router
    router = slice.add_node(name="router", site = site2, cores=CORES, ram=RAM, image=IMAGE, disk = 50)
    r_h1_iface = router.add_component(model='NIC_Basic', name="if_r_h1").get_interfaces()[0]
    r_h2_iface = router.add_component(model='NIC_Basic', name="if_r_h2").get_interfaces()[0]
    
    #Create h1 router
    h1 = slice.add_node(name="h1", site = site2, cores=CORES, ram=RAM, image=IMAGE, disk = 20)
    h1_iface = h1.add_component(model='NIC_Basic', name="if_h1").get_interfaces()[0]
    
    #Create h2 router
    h2 = slice.add_node(name="h2", site = site2, cores=CORES, ram=RAM, image=IMAGE, disk = 20)
    h2_iface = h2.add_component(model='NIC_Basic', name="if_h2").get_interfaces()[0]
    
    h1_net = slice.add_l2network(name="h1_net", interfaces=[r_h1_iface, h1_iface])
    h2_net = slice.add_l2network(name="h2_net", interfaces=[r_h2_iface, h2_iface])
except Exception as e:
    print(f"Exception: {e}")
```

```python
slice.submit()
```

```python
now = datetime.datetime.now(datetime.timezone.utc)
end_date = (now + timedelta(days=14)).strftime("%Y-%m-%d %H:%M:%S %z")
slice.renew(end_date)
```

```python
from ipaddress import ip_address, IPv4Address, IPv6Address, IPv4Network, IPv6Network

try:
    subnet = IPv4Network("192.168.1.0/24")
    available_ips = list(subnet)[1:]
    print(available_ips)
except Exception as e:
    print(f"Exception: {e}")
```

```python
router_node = slice.get_node("router")
router_iface = router_node.get_interface(network_name="h1_net") 
ip = available_ips.pop(0)
router_iface.ip_addr_add(addr=ip, subnet=subnet)
```

```python
stdout, stderr = router_node.execute(f'ip addr show {router_iface.get_os_interface()}')
print (stdout)
```
