```python
import json
import traceback

from fabrictestbed_extensions.fablib.fablib import fablib
```

```python
SLICE_NAME = 'nginx-lb-slice'
SITE="UCSD"
NUM_SERVERS = 3

# We will use Ubuntu 20.04 for both nodes
IMAGE = 'default_ubuntu_20'
CORES = 1
RAM = 2
server_names = ["server_{}".format(i) for i in range(NUM_SERVERS)]
print(server_names)
```

```python
try:
    slice = fablib.get_slice(name=SLICE_NAME)
    for node in slice.get_nodes():
        print(f"{node}")
except Exception as e:
    print(f"Exception: {e}")
```

```python
for server in server_names:
    try:
        print("Installing Flask on server {}".format(server))
        server_node = slice.get_node(server)
        stdout, stderr = server_node.execute("python3 --version")
        print(stdout, stderr)
        print(server_node.execute("sudo apt-get update && sudo apt install -y python3-pip"))
        print(server_node.execute("pip3 --version"))
        stdout, stderr = server_node.execute("pip3 install Flask")
        print(stdout, stderr)
        print("Finished installing Flask.")

    except Exception as e:
        print(f"Exception: {e}")
```
